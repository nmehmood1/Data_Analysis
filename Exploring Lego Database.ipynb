```python
# Short project demonstrating use of pandas dataframes to work with relational database using grouping and sorting.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

# Read colors data
colors = pd.read_csv('lego\colors.csv')
num_colors = colors['name'].size
print("number of colours: " + str(num_colors))

# colors_summary: Distribution of colors based on transparency
colors_summary = colors.groupby('is_trans').count()
#print(colors_summary)

%matplotlib inline
# Read sets data as `sets`
sets = pd.read_csv('lego\sets.csv')

# summary of average number of parts by year
parts_by_year = sets[['year','num_parts']].groupby('year',as_index=False).mean()
#print(parts_by_year)

# Plot trends in average number of parts by year
parts_by_year.plot('year','num_parts')
plt.title('Average Number of Parts vs Year')
plt.xlabel('Year')
plt.ylabel('Average Number of Parts')


# themes_by_year: Number of themes shipped by year

themes_by_year = sets[['year','theme_id']].groupby('year',as_index=False).count()
#print(themes_by_year.head(2))
themes_by_year.plot('year','theme_id')
plt.title('Number of Themes Shipped vs Year')
plt.xlabel('Year')
plt.ylabel('Number of Themes')

# What is the set with the highest number of parts?

sets_SBP = sets[['name','num_parts']].sort_values('num_parts',ascending=False)
print('Sets with the highest number of parts:')
print(sets_SBP.head(3))
```
